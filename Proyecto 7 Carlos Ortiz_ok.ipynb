## Prueba de hipótesis

### Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

**Prueba de levene**
+ h0 = la varianza de de las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las iguales.

+ h1 = la varianza de de las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son diferentes.


```python
ing_1= df_2017[df_2017["platform"]=="XOne"]["user_score"].dropna()
ing_2= df_2017[df_2017["platform"]=="PC"]["user_score"].dropna()

alpha = 0.05

var= st.levene(ing_1, ing_2)

print('Valor p:', var.pvalue)

if (var.pvalue<alpha):
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

Según el valor p obtenido en la prueba de Levene, que se encuentra significativamente por debajo del nivel de significancia establecido (0.05), podemos concluir que hay suficiente evidencia para rechazar la hipótesis nula. Esto indica que la varianza de `user_score` para "XOne" y "PC" no son iguales. En otras palabras, existe una diferencia significativa en la variabilidad de los scores de las dos plataformas, lo que sugiere que el comportamiento de  pueden variar considerablemente.

**Prueba de hipotesis**

+ h0 = Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.
+ h1 = Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son diferentes.

```python
ing_1= df_2017[df_2017["platform"]=="XOne"]["user_score"].dropna()
ing_2= df_2017[df_2017["platform"]=="PC"]["user_score"].dropna()

alpha = 0.05

results = st.ttest_ind(ing_1,ing_2,equal_var=False)

print('Valor p:', results.pvalue)

if (results.pvalue<alpha):
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
    
```
